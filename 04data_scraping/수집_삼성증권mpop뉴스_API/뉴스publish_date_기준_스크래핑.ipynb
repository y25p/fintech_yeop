```python
import requests
import pandas as pd
from datetime import datetime

def json_unpack(json_data):
    items = json_data['items']
    df = pd.json_normalize(items)
    return df

# 🔍 검색어 및 날짜 범위 설정
keyword = "mpop"
start_date = "2025-01-22"
end_date = "2025-05-24"
start_dt = datetime.strptime(start_date, "%Y-%m-%d").date()
end_dt = datetime.strptime(end_date, "%Y-%m-%d").date()

# 🔐 API 인증
Client_Id = "sk2iKdMANqpvG3Pa2WIJ"
Client_Secret = "by6xHZcrlB"
headers = {
    "X-Naver-Client-Id": Client_Id,
    "X-Naver-Client-Secret": Client_Secret
}

# 📦 데이터 수집
all_data2 = []
start = 1

while True:
    url = "https://openapi.naver.com/v1/search/cafearticle.json"
    payload = dict(query=keyword, display=100, start=start, sort="sim")
    r = requests.get(url, params=payload, headers=headers)
    print(r.url)
    print("Status:", r.status_code)

    if r.status_code != 200:
        break

    data = r.json()
    if not data.get('items'):
        break

    result_df = json_unpack(data)

        # 🗓 pubDate 파싱 및 날짜 필터링
    if 'pubDate' in result_df.columns:
        result_df['pubDate'] = pd.to_datetime(
            result_df['pubDate'], 
            format='%a, %d %b %Y %H:%M:%S %z', 
            errors='coerce'
        )
        result_df['pubDate_only'] = result_df['pubDate'].dt.date

        # ✅ 중간 점검 출력
        print("✔️ 날짜 파싱 확인:", result_df[['title', 'pubDate_only']].head(3))

        # ✅ 날짜 비교 전에도 중간 값 출력
        print("🔍 필터 기준:", start_dt, "~", end_dt)

        result_df = result_df[
            (result_df['pubDate_only'] >= start_dt) & 
            (result_df['pubDate_only'] <= end_dt)
        ]
    else:
        print("❌ pubDate가 없음")
        result_df = pd.DataFrame()
        
    # 저장 직전 확인
if all_data2:
    final_result2 = pd.concat(all_data2, ignore_index=True)
    print("🔎 최종 수집된 pubDate 범위:", final_result2['pubDate_only'].min(), "~", final_result2['pubDate_only'].max())


# 🧾 최종 CSV 저장
if all_data2:
    final_result2 = pd.concat(all_data2, ignore_index=True)
    final_result2.to_csv("네이버검색_카페_mpop.csv", encoding="utf-8-sig", index=False)
    print("✅ 저장 완료! 수집 기사 수:", len(final_result2))
else:
    print("⚠️ 지정한 날짜 범위에 해당하는 뉴스가 없습니다.")
```
