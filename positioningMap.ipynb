```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
```

```python

# 3페이
pays = ['naver', 'kakao', 'toss']

# 페이의 적립 (X축)
point = [5, 3, 2]

# 각 페이의 성장율 (Y축)
shopping_MAU = [5, 2, 3]

# 포지셔닝 맵 생성
plt.figure(figsize=(8, 6))
plt.scatter(point, shopping_MAU, color='skyblue', s=100)

# 각 점에 브랜드 이름 표시
for i in range(len(pays)):
    plt.text(point[i]+0.2, shopping_MAU[i]+0.2, pays[i], fontsize=10)

# 축 라벨 및 제목
plt.xlabel('point (적립)')
plt.ylabel('shopping_MAU (성장율)')
plt.title('Korea Pay Apps Positioning Map')

# 축 범위 조절 (선택)
plt.xlim(0, 10)
plt.ylim(0, 10)

# 그리드 및 표시
plt.grid(True)
plt.axhline(y=5, color='gray', linestyle='--', linewidth=1)
plt.axvline(x=5, color='gray', linestyle='--', linewidth=1)
plt.show()
```
