```python
# 2    10 ?
second= np.array([[1,2,3],[1,2,3]])
print(second, second.ndim, second.shape)
```

```python
for num in second:
    for number in num:
        print(number*10) #  10  
```

```python
result=[] #     append
for num in second:
    mid_result=[]
    for number in num:
        mid_result.append(number*10) 
    result.append(mid_result)
result=np.array([result])
print(result.ndim, result.shape, result)
```

# .argsort()     
    .


```python
g= np.array([5, 0.6, -7, 42, 10, 0]) #   [2] -> [5] -> [1]  
print(g)
```

```python
s=g.argsort()
print(s, s.dtype)
```

```python
print(g[s])
print(g[s][::-1])
```
