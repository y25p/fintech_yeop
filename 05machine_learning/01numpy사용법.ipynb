```python
# 1차원 (벡터)의 연산
b1=[1,2,3,4,5]

# 리스트의 요소에 10을 곱한 후 다시 리스트로.
# b1*10 # 이렇게하면 리스트가 10번씩 반복됨.

result=[]
for num in b1:
    result.append(num*10)
result
```

```python
# 2차원 데이터도 모든 요소에 10을 곱해야한다면?
second= np.array([[1,2,3],[1,2,3]])
print(second, second.ndim, second.shape)
```

```python
for num in second:
    for number in num:
        print(number*10) # 여기서 10곱해야 하는 상황
```

```python
result=[] # 원상태로 되돌리려면 또 변수에 append
for num in second:
    mid_result=[]
    for number in num:
        mid_result.append(number*10) 
    result.append(mid_result)
result=np.array([result])
print(result.ndim, result.shape, result)
```

```python
#3차원 데이터요소에 10 곱하기.
third= np.array([[[1,2,3],[1,2,3]],[[1,2,3],[1,2,3]]])
print(third, third.ndim, third.shape)
```

```python
for num1 in third:
    print(num1, end="\n\n") # 아직도 리스트형태
```

```python
for num1 in third:
    for num2 in num1:
        print(num2, end="\n\n") # 아직 리스트형태
```

```python
for num1 in third:
    for num2 in num1:
        for num3 in num2:
            print(num3, end="\n\n") # 드디어 요소. 이제 곱하면 됨.
```

```python
result=[]
for num1 in third:
    num1_result=[]
    for num2 in num1:
        num2_result=[]
        for num3 in num2:
            num2_result.append(num3 * 10) # 드디어 요소. 이제 곱하면 됨.
        num1_result.append(num2_result)
    result.append(num1_result)
result=np.array(result)
result
```

## 정렬시켜주는 함수
- .sort(), .sorted()
- .sort(값, reverse=True) .sorted(값, reverse=True)
- NumPy의 .sort()에는 reverse=True 옵션이 없다.
- 역순 정렬은 np.sort(arr)[::-1] 또는 np.argsort(arr)[::-1]을 사용!
- .sort()는 배열 자체를 바꾸고, np.sort()는 복사본 반환함
- 일반 파이썬은 .sort(reverse=True)와 sorted(..., reverse=True) 둘 다 지원.

```python
f= np.array([5,7,8,2,4,23,0.3])
f
```

```python
f.sort() #sort는 덮어씀
f
```

```python
sorted(f) # 재할당해야하는 것 헷갈림
```

```python
print(sorted(f,reverse=True)) # 역순 공식~
f
```

```python
f.sort(reverse=True)
f
```

# .argsort() 자료값이 오름차순이 되도록 인덱스 출력
사용자 선호에따른 추천시스템을 만들때 쓰임.


```python
g= np.array([5, 0.6, -7, 42, 10, 0]) # 오름차순으로 정렬하려면 [2] -> [5] -> [1] 요런식으로 가야함
print(g)
```

```python
s=g.argsort()
print(s, s.dtype)
```

```python
print(g[s])
print(g[s][::-1])
```
