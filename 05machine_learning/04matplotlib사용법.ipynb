```python
x=np.linspace(0,2,100) # 100  0~2 
x
```

```python
# OO-style
fig, ax=plt.subplots(figsize=(5, 2.7))
ax.plot(x,x,label='linear')
ax.plot(x,x**2, label="quadratic")
ax.plot(x,x**3, label="cubic")
ax.set_xlabel('x label') #x 
ax.set_ylabel('y label') #y 
ax.set_title(' ') # 
ax.legend() # 
plt.show()
```

```python
#          3 1
fig, ax=plt.subplots(3, 1,figsize=(5, 2.7))
#
ax[0].plot(x,x,label='linear')
ax[1].plot(x,x**2, label="quadratic")
ax[2].plot(x,x**3, label="cubic")
ax[0].set_xlabel('x label') #x 
ax[0].set_ylabel('y label') #y 
ax[0].set_title(' ') # 
ax[0].legend() # 
plt.show()
```
