# matplotlib 2가지 코딩스타일<br>(1)OO스타일
- 표준 권장 코딩방법.
- plt(pyplot)보다 복잡한 사용법
- 복잡한 그래프를 그릴 때 씀.

```python
x=np.linspace(0,2,100) # 100개의 구간으로 0~2를 쪼개라
x
```

```python
# OO-style
fig, ax=plt.subplots(figsize=(5, 2.7))
ax.plot(x,x,label='linear')
ax.plot(x,x**2, label="quadratic")
ax.plot(x,x**3, label="cubic")
ax.set_xlabel('x label') #x축과 그이름
ax.set_ylabel('y label') #y축과 그이름
ax.set_title('간단 그래프') # 그래프이름
ax.legend() # 범례
plt.show()
```

```python
# 따로 그리기        3행 1열
fig, ax=plt.subplots(3, 1,figsize=(5, 2.7))
#인덱스추가
ax[0].plot(x,x,label='linear')
ax[1].plot(x,x**2, label="quadratic")
ax[2].plot(x,x**3, label="cubic")
ax[0].set_xlabel('x label') #x축과 그이름
ax[0].set_ylabel('y label') #y축과 그이름
ax[0].set_title('간단 그래프') # 그래프이름
ax[0].legend() # 범례
plt.show()
```

# (2) pyplot 스타일
- 그래프 하나 빨리 그려야할때 OO스타일보다 간단함
- 복잡한 그래프에는 부적합
- OO-스타일과비교<br>

- EDA하는 방법은 여러그래프를 동시에 비교하기보다 하나씩 보게되므로 <br> pyplot이 더 많이 쓰이게됨

```python
# pyplot스타일
plt.figure(figsize=(5, 2.7))
plt.plot(x,x,label='linear')
plt.plot(x,x**2, label="quadratic")
plt.plot(x,x**3, label="cubic")
plt.xlabel('x축이름') 
plt.ylabel('y축이름') 
plt.title('pyplot스타일 그래프') # 그래프이름
plt.legend() # 범례
plt.show()
```

# scatter 그래프

**cmyk로 색주기**

```python
fig, ax= plt.subplots(figsize=(5,2.7))    # 색변경 부분은 구글링해도 충분.
ax.scatter(data1, data2, s=100, facecolor='C2', edgecolor='b')
plt.show()               # 점크기
```

# 마커의 모양 지정

| 마커 | 설명 |
