```python
import gradio as gr
import google.generativeai as genai

# 🔐 API 키 설정
genai.configure(api_key=apiikey)  # ← 여기에 발급받은 API 키 넣기

# 📘 모델 선택
model = genai.GenerativeModel("gemini-2.0-flash")

# 🤖 Gemini 응답 함수
def chat_with_gemini(user_input, history):
    # Gradio Chatbot 컴포넌트와 연동되는 대화 이력 처리
    history = history or []
    prompt = "\n".join([f"사용자: {msg[0]}\nAI: {msg[1]}" for msg in history])
    prompt += f"\n사용자: {user_input}\nAI:"

    # Gemini API 호출
    try:
        response = model.generate_content(user_input)
        answer = response.text.strip()
    except Exception as e:
        answer = f"⚠️ 오류 발생: {e}"

    history.append((user_input, answer))
    return history, history

# 🎨 Gradio UI 구성
with gr.Blocks() as demo:
    gr.Markdown("## 💬 Gemini 2.0 Flash 챗봇")

    chatbot = gr.Chatbot(label="Gemini와 대화하기")
    msg = gr.Textbox(label="질문을 입력하세요", placeholder="예: 시험기간을 표현한 시를 써줘", lines=2)
    send_btn = gr.Button("보내기")

    state = gr.State([])  # 대화 이력 저장

    send_btn.click(fn=chat_with_gemini, inputs=[msg, state], outputs=[chatbot, state])
    msg.submit(fn=chat_with_gemini, inputs=[msg, state], outputs=[chatbot, state])  # Enter로도 전송 가능

# 🚀 실행
demo.launch()
```

```python
# 뉴스 기사 본문
import gradio as gr
import pandas as pd
import requests
from bs4 import BeautifulSoup
import google.generativeai as genai

# ✅ 1. API 키 및 모델 설정
api_key = "AIzaSyDjxtCeoSu6jrVEkExDc5RZn1hPJ31OLoE"  # 실제 Google API 키로 교체하세요
genai.configure(api_key=api_key)
model = genai.GenerativeModel("models/gemini-2.0-flash")

# ✅ 2. 뉴스 검색 함수 (네이버 API)
def fetch_news_titles(query):
    url = "https://openapi.naver.com/v1/search/news.json"
    headers = {
        "X-Naver-Client-Id": "sk2iKdMANqpvG3Pa2WIJ",
        "X-Naver-Client-Secret": "by6xHZcrlB"
    }
    params = {"query": query, "display": 10, "start": 1, "sort": "sim"}
    res = requests.get(url, params=params, headers=headers)
    items = res.json().get("items", [])
    titles = [item["title"].replace("<b>", "").replace("</b>", "") for item in items]
    links = [item["link"] for item in items]
    return pd.DataFrame({"제목": titles, "링크": links})

# ✅ 3. 기사 본문 추출
def extract_article_text(url):
    try:
        res = requests.get(url, timeout=5)
        soup = BeautifulSoup(res.text, "html.parser")
        paragraphs = soup.find_all("p")
        text = "\n".join([p.get_text().strip() for p in paragraphs if len(p.get_text()) > 50])
        return text if len(text) > 200 else "기사 본문을 충분히 추출할 수 없습니다."
    except Exception as e:
        return f"본문 추출 오류: {e}"

# ✅ 4. 쉬운 말 Gemini 요약 함수
def summarize_for_kids(text):
    prompt = (
        "다음은 한 뉴스 기사의 본문입니다.\n\n"
        "이 내용을 12살 어린이가 이해할 수 있도록 **쉽게 설명해주되 3-4문장으로 요약해줘.**\n"
        "뉴스 섹션, 추천, 댓글 정책, 프리미엄 콘텐츠 같은 부가 설명은 포함하지 마.\n\n"
        f"{text}"
    )
    try:
        response = model.generate_content(contents=prompt)
        return response.text
    except Exception as e:
        return f"Gemini 오류: {e}"

# ✅ 5. 상태 관리
state_df = pd.DataFrame()

def update_dropdown(query):
    global state_df
    state_df = fetch_news_titles(query)
    return gr.update(choices=state_df["제목"].tolist()), ""

def summarize_selected_title(selected_title):
    global state_df
    link = state_df[state_df["제목"] == selected_title]["링크"].values[0]
    article = extract_article_text(link)
    if article.startswith("본문 추출 오류") or article.startswith("기사 본문을"):
        return article
    return summarize_for_kids(article)

# ✅ 6. Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("## 🧒 어린이용 뉴스 요약기 (Gemini 2.0 Flash 기반)")

    with gr.Row():
        with gr.Column(scale=1):
            keyword_input = gr.Textbox(label="검색어 입력", placeholder="예: 환경, 과학, 경제")
            search_btn = gr.Button("뉴스 검색")
            title_dropdown = gr.Dropdown(label="뉴스 제목 선택", choices=[])
        with gr.Column(scale=2):
            summary_output = gr.Textbox(label="Gemini 요약 결과 (어린이용)", lines=25)

    search_btn.click(fn=update_dropdown, inputs=keyword_input, outputs=[title_dropdown, summary_output])
    title_dropdown.change(fn=summarize_selected_title, inputs=title_dropdown, outputs=summary_output)

demo.launch()
```
