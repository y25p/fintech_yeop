## 왼쪽, 오른쪽, 양쪽 공백/특정 문자 제거하기 
- lstrip() : 지정한것을 왼쪽만 삭제
- rstrip() : 지정한 문자를 문자열 오른쪽에서 삭제
- strip() : 문자열 앞뒤 양쪽에서 지정한 문자 찾아 삭제

```python
h8='       공백 있는 문자열'
h8.lstrip()
```

```python
h8='   공백 있는 문자열    '
h8.rstrip()
```

- 글자가 있는 부분은 지정되지 않았으므로 그부분 부터는 지정한 내용이 사이에 껴있어도 지우지 못함.
- 그런경우에도 지워야한다면 strip과 replace 같이 쓸것.

```python
h8.strip().replace(" ","")
```

```python
h8 #양옆에 각각 3, 4칸씩 공백인 문자열
```

# 정규표현식
- 패턴에 맞는 자료들만 추출 가능
- 더욱 간단해짐

```python
# re (정규표현식) 불러오기
import re
```

```python
# 예제
text='안녕하세요! this is a test. 1234 !@$$@#()%$^&%^*'
re.sub(r'[^가-힣a-zA-Z0-9]','',text) # sub메서드는 대체기능
```

```python
re.sub(r'[^가-힣a-zA-Z0-9]',' ',text) #공백주면
```
