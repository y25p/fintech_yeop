# if else 조건문
## 특정조건 판별하여 작업 분기하는 문법

- if조건식 : 
- __4칸_조건이 참일 때 실행할 코드
- else: __4칸_조건이 거짓일때 실행할 코드

```python
x=13
if x==10: #콜론쳐야 다음 줄을 4칸띄게 잡아줌
    print(f'x는, {x}입니다.')
else:
    print(f'{x}는, 10이 아닙니다.')
```

## elif 이엘이프 다중조건문
- if 조건식:<br>
____조건이 참일 때 실행할 코드<br>
- 첫 조건식이 거짓일 경우 else로 떨어지는데 그사이에<br>
- elif를 삽입해 2번째 조건 넣기 가능<br>
elif 조건식2:<br>
____조건식2가 참일 때 실행할 코드<br>
elif 조건식3:<br>
_____조건식3이 참일 때 실행할 코드<br>
else:<br>
_____조건식 1,2,3이 모두 거짓일 때 실행할 코드<br>

- 점수가 81-100이내면 A, 61-80이면 B, 41-60은 C, 21-40은 D, 0-20까지가 F로 분류하는 조건식을 만들어보려면

```python
score=int(input('점수 입력하세요'))
if 0 <= score <101:
    if 100>= score >80 :
        print(f'{score}는 A등급')
    elif 81> score > 60:
        print(f'{score}는 B등급')
    elif 61> score > 40:
        print(f'{score}는 C등급')
    elif 41> score > 20:
        print(f'{score}는 D둥급')
    elif 21> score > 0:
        print(f'{score}는 F등급')
else :  #else 조건은 없다는점 헷갈리지말것
    print(f'{score}는 분석할 수 없습니다. 점수를 정확하게 입력하세요')
```

### 문제. 1-50사이 숫자를 입력받아 3의 배수면 '3의배수입니다.' 3의배수가 아니면 '3의 배수가 아닙니다.' 
1. 입력받기
2. 정수로 변환하기
3. 1-50사이 숫자인가?
4. 3의 배수인가?
5. 출력

```python
susja= int(float(input('숫자를 입력하세요')))
if 50>susja >1 :
    if susja % 3 ==0 :
        print(f'{susja}는 3의 배수입니다.')
    elif susja % 3 !=0 :
        print(f'{susja}는 3의 배수가 아닙니다.')
else:
    print(f'{susja}는 범위에 속하지 않으니 다른 수를 입력해보세요.')
```

```python
susja= int(float(input('숫자를 입력하세요'))) #type(susja) 를 해서 dt꼭확인
if 50>susja >1 :
    if susja % 3 ==0 :
        print(f'{susja}는 3의 배수입니다.')
    else:
        print(f'{susja}는 3의 배수가 아닙니다.')
else:
    print('1~50사이 숫자를 입력해보세요.')
```

## 3. 조건표현식
- 파이썬에만있는 조건표현식.
- if/else문
- 참일 때 실행코드 if 조건, else 거짓일 때 실행 코드

```python
num=int(input('숫자를 입력하시오'))
print(num, type(num))
```

```python
num=int(input('숫자를 입력하시오'))
if num % 2 == 0:
    print('짝수')
else:
    print('홀수') # 이렇게 아무리 못해도 4줄을 써야한다. 일반적 if else 문.
```

```python
#위 내용을 한줄로
num=int(input('숫자를 입력하시오')) 
print('짝수') if num % 2 == 0 else print('홀수')
```

```python
# list나 dict, for문도 섞어서 한줄 가능하다.
num=int(input('숫자를 입력하시오')) 
print('짝수') if num % 2 == 0 else print('홀수')
```
