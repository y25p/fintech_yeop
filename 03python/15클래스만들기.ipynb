````python
#데이터입력을 잘받으니 함수를 더만들자.
class FourCal():
    def setdata(self, num1, num2): #*붙이면 튜플로받아서 계산안됨.
        self.num1=num1
        self.num2=num2
        print("self.num1",self.num1)
        print("self.num2",self.num2)
        
    #덧셈
    def add(self):
        result =self.num1+self.num2 #누구자료인지 알수있게
        return result               #self가 따라다님
    #뺼셈
    def sub(self):
        result=self.num1-self.num2
        return result
    #곱
    def mul(self):
        result=self.num1*self.num2
        return result
    #/
    def div(self):
        result=self.num1/self.num2
        return result
````

````python
sam=FourCal()
sam.setdata(1,1)
sam.add()
````

````python
sam.setdata(3,5)
sam.div()
````

# 생성자 (constructor)
- 클래스를 실행해서 인스턴스를 만들때 초기값을 받도록해줌
- 생성자는 인스턴스(객체) 생성시 자동으로 호출<br>
주로 초기값을 받을때 씀
- 생성자 만드는법<br> 클래스 내에서 매서드명으로<br>`__init__`를 사용


- 이렇게 생성되면 아래의 다른함수(add)에서도 변수를 받게됨.
- 생성자를 이용해 초기값을 입력받는 FourCal2를 만들기.

````python
class FourCal2():
    def __init__(self, num1, num2): #인스턴스생성할때 같이num1,2
        self.num1=num1
        self.num2=num2
    def add(self):
        result = self.num1+self.num2
        return result
    def sub(self):
        result = self.num1-self.num2
        return result
    def mul(self):
        result = self.num1*self.num2
        return result
    def div(self):
        result = self.num1/self.num2
        return result
        
        
````

### 이런 제로디비전에러를 수정하기위해<br> SafeFourCal을 만들고 FourCal2를 상속시켜서 <br>오류난 부분만 재정의하기.

````python
class SafeFourCal(FourCal2): #상속받았고
    def div(self):           #업데이트할부분.
        if self.num2 ==0:
            return 0
        else:
            return self.num1/self.num2
````

## 메서드 오버로딩

- 동일한 메서드가 기능만 변경된채 여러개로 존재<br>
`def add(num1,num2)`<br>
`def add(num1,num2,num3)`<br>
`def add(num1,num2,num3, num4)`<br>



````python
class OverloadingEx():
    def add(self, x, y):
        return x+y
    def add(self, x, y, z):
        return x+y+z
    def add(self, x, y, z, a): #이부분만 작동하도록 된 상태.
        return x+y+z+a
    
    
````

### multipledispatch 패키지(라는 모듈)를 통해서 오버로딩 구현

````python
# !pip install multipledispatch #설치 후 주석처리
````

````python
from multipledispatch import dispatch
````

````python
class OverloadingEx2():
    @dispatch(int,int)
    def add(self, x, y):
        return x+y
    @dispatch(int,int,int)
    def add(self, x, y, z):
        return x+y+z
    @dispatch(int,int,int,int)
    def add(self, x, y, z, a):
        return x+y+z+a
    @dispatch(float, float,float,float)
    def add(self, x, y, z, a):
        return x+y-z*a
````

````python
sam=OverloadingEx2()
sam.add(2,3)
````

````python
sam.add(2,3,3)
````

````python
sam.add(2.1,3.1,3.0,2.4) 
````

- 인스턴스에따라다르게 기억저장관리할 자료가 아니면 쓰기편함.


````python
#self의 흔적이없음.
class Cal():
    @staticmethod
    def add(a,b):
        print(a+b)
        
    @staticmethod
    def sub(a,b):
        print(a-b) #단순계산이지 결과물을 저장하지않음
````

````python
sam=Cal()
sam.add(4,5) 
````

````python
sally=Cal()
sally.add(10,3)
````

````python
#인스턴스 생성도 안하고 바로쓰기
Cal.add(9,10)
````

# 클래스 메서드 class method
`@classmethod`라는 데코레이터를 붙임.
- 클래스 메서드의 첫 매개변수는 cls지정(필수)
- 인스턴스 생성없이 사용가능.
- 메서드 안에서 `클래스속성, 클래스메서드`에 접근해야할때 사용.

````python
class Factory():
    n_coffee_machine=0 #클래스 속성
    
    def __init__(self):
        Factory.n_coffee_machine +=1 #인스턴스가 생성될 때 
                                  #클래스 속성에 1을 더하라는뜻.
            
    @classmethod
    def print_n_coffee_machine(cls):
        print(f"{cls.n_coffee_machine}개가 제조되었습니다.")
        
#인스턴스 생성하기 전        
````

````python
Factory.print_n_coffee_machine() #찍어보면 0개 제조로 나옴.
````

````python
# 인스턴스 생성해보기
sally=Factory()
sally.print_n_coffee_machine() #생성하고나니까 숫자가 올라가네...
````

# 📌추상 클래스 abc(absctract base class
- 혼자서는 만들수없고 abc 모듈이 필요함.

- 추상클래스는 클래스를 상속받아 만들 때 <br>반드시 만들어야 하는 메서드를 지정해주는 클래스
- 인터페이스라고도하며
- 규칙에따라 클래스를 만들도록 강제함.
- 추상클래스에서 정의한 메서드를 상속할때 반드시 Overriding.<br> 추상클래스에서 정의한대로.
- 리모컨의 버튼들마다 기능이 정해져있으니 <br> 매뉴얼대로만 만들면됨.

````python
from abc import*
````

````python
# 추상클래스
class CoffeeBase(metaclass=ABCMeta):
    @abstractmethod
    def bean_input(self):
        pass

    @abstractmethod
    def grind(self):
        pass
    
````

````python
sam=CoffeeBase()

#결과는 에러.
#추상클래스는 인스턴스 생성 불가
````

##### 그대로 오버라이딩하지 못하고 일부만 한다면?


````python
class CoffeeMachine(CoffeeBase):
    def bean_input(self):       #이부분말고도 grind도 def되어야함
        print('커피 원두 넣기')
````

````python
sam = CoffeeMachine() #인스턴스 생성이 안되는 이유 ) grind도 def되어야함
````

````python
class CoffeeMachine(CoffeeBase):
    def bean_input(self): 
        print('커피 원두 넣기')
        
    def grind(self):
        print('커피 원두 갈기')
````

````python
sam = CoffeeMachine() 
sam.grind()
````
